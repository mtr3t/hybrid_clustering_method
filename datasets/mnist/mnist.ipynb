```python
f = open('01_mnist.ds', 'w')
for i in range(10):
    images = mnist['train' + str(i)]
    for j in range(len(images)):
#     for j in range(1):
        temp = images[j]
        f.write(str(temp).strip('[]').replace('\n','').lstrip('  '))
        f.write('\n')
f.close()
```

```python
f = open('01_mnist_ground_truth.ds', 'w')
for i in range(10):
    images = mnist['train' + str(i)]
    for j in range(len(images)):
        f.write(str(i))
        f.write('\n')
f.close()
```
