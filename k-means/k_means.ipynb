```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import normalized_mutual_info_score
```

```python
def k_means(problem, ground_truth, plotter=False, printer=False):
    """perform kmeans clustering
    keyword arguments:
        problem: to cluster
        ground_truth: of the problem
        nclusters: number of cluster to cluster by
        plotter: plot the problem default=False
        printer: print out statments for debugging default=False
    returns:
        normalized mutual information score
    """
    # load the problem
    X = np.loadtxt(problem)
    if printer: print('the shape of the problem is:', X.shape)
    
    # load the ground truth
    gt = np.loadtxt(ground_truth).astype(np.int32)
    if printer: print('the ground truth is: ', gt)
        
    # calculate the nuber of clusters from the ground truth
    nclusters = len(np.unique(gt))
    if printer: print('number of clusters =', nclusters)
        
    kmeans = KMeans(n_clusters=nclusters, n_init=10).fit(X)
    if printer: print('k-means labels are: ', kmeans.labels_)
        
    # calculate the normalized mutual information score
    nscore = round(normalized_mutual_info_score(gt, kmeans.labels_), 3)*100
    print(problem, '\nNormalized Mutual Information Score:', nscore, '%')
    
    if plotter:
        if X.shape[1] == 2:
            import matplotlib.pyplot as plt
            %matplotlib inline 
            
            # print out problem
            plt.scatter(X[:,0], X[:,1], color = [["red", "blue"][i] for i in kmeans.labels_])
            plt.title('K-means: ' + problem + '\nNormalized Mutual Info Score: ' + str(nscore) + '%')
            plt.ylabel('y')
            plt.xlabel('x')
```
